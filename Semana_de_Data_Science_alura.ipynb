#Machine Learning

```
from sklearn.model_selection import train_test_split
```

```
provas_entrada = ['NU_NOTA_CH','NU_NOTA_LC','NU_NOTA_REDACAO','NU_NOTA_CN']
provas_saida= 'NU_NOTA_MT'

dados_sem_notas_zero = dados_sem_notas_zero[provas].dropna()#precisamos tratar os NaN para criar um modelo de machine learning
#neste caso, eu retirei as pessoas que não fizeram algumas provas

notas_entrada = dados_sem_notas_zero[provas_entrada]
notas_saida = dados_sem_notas_zero[provas_saida]
```

```
notas_entrada
```

```
x= notas_entrada
y= notas_saida
```

```
seed = 7

x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size= 0.25, random_state= seed)
```

```
x_treino.head()
```

```
predicoes_matermatica = modelo.predict(x_teste)
```

```
plt.figure(figsize= (8,8))
sns.scatterplot(x=predicoes_matermatica, y= y_teste)
plt.xlim((-50,1050))
plt.ylim((-50,1050))
```

#Validação de modelo e Overfit

```
modelo = LinearSVR(random_state=seed)
modelo.fit(x_treino, y_treino)
predicoes_matermatica = modelo.predict(x_teste)
mean_squared_error(y_teste, predicoes_matermatica)
```

```
#Testando a arvore de decisão como modelo
from sklearn.tree import DecisionTreeRegressor

modelo_arvore = DecisionTreeRegressor(max_depth= 3)
modelo_arvore.fit(x_treino, y_treino)
precicores_matematica_arvore = modelo_arvore.predict(x_teste)
```

```
mean_squared_error(y_teste, precicores_matematica_arvore)
#a árvore de decisão se saiu melhor do que o linearSVM
```

```
#implementando cross validation para melhorar na seleção de dados para treino
from sklearn.model_selection import cross_validate

modelo_arvore= DecisionTreeRegressor(max_depth=3)
resultados= cross_validate(modelo_arvore, x, y, cv= 10,scoring= "neg_mean_squared_error")#cv é a quantidade de validação

#a gente ta usando o mean squared erro como metrica de avalição, mas pra ser aceito no cross validade, precisar ser o neg mean squared erro
#quanto mais próximo de zero é melhor

media = (resultados["test_score"]*-1).mean()
desvio_padrao = (resultados["test_score"]*-1).std()
lim_inferior = media -(2*desvio_padrao)
lim_superior = media + (2*desvio_padrao)

print(f"Intervalo de confiança: {lim_inferior} - {lim_superior}")
```

```
#Criando uma função pra facilitar
def calcula_mse(resultados):
  media = (resultados["test_score"]*-1).mean()
  desvio_padrao = (resultados["test_score"]*-1).std()
  lim_inferior = media -(2*desvio_padrao)
  lim_superior = media + (2*desvio_padrao)

  print(f"Intervalo de confiança: {lim_inferior} - {lim_superior}")
```

```
from sklearn.model_selection import KFold
#Kfold é pra embaralhar e dividir os dados na hora de selecionar os dados (embaralhar= shuffler/ dividir = n_split)
#pois pode acontecer na hora de selecionar pro treino, pegar mt dados de um grupo e menos do outro

SEED = 123
np.random.seed(SEED)#tudo que estiver nesse célula vai usar o SEED criado

partes = KFold(n_splits = 10, shuffle=True)
modelo_arvore= DecisionTreeRegressor(max_depth=3)
resultados= cross_validate(modelo_arvore, x, y, cv= partes,scoring= "neg_mean_squared_error")

calcula_mse(resultados)
```

```
#Criar uma função pra escolher quantas max_depth fica melhor no nosso modelo
def regressor_arvore(nivel):
  SEED = 123
  np.random.seed(SEED)
  partes = KFold(n_splits = 10, shuffle=True)
  modelo_arvore= DecisionTreeRegressor(max_depth=nivel)
  resultados= cross_validate(modelo_arvore, x, y,
                             cv= partes,scoring= "neg_mean_squared_error", return_train_score=True)
  print(f'Treino = {(resultados["train_score"]*-1).mean()} | Teste = {(resultados["test_score"]*-1).mean()}')

regressor_arvore(4)
```

```
for i in range(1,21):
  print(f'i valendo:{i}')
  regressor_arvore(i)
  print('')

 #os treinos vão diminuindo e o teste aumentando porque o nosso modelo fica tao bom nos treinos e acaba decorando so valores...
 #e na hora de generalizar pro teste, ele começa a dar erro. é o tal do overfit 
```
